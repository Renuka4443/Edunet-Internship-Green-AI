```python
#Creating the date range
date_range=pd.date_range(start="2025-01-01",end="2025-10-01",freq="MS")
date_range
```

```python
#extracting the components of dates
df=pd.DataFrame(date_range)
df
```

```python
df['Date'] = df[0]
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df
```

```python
df.drop(0,axis=1,inplace=True)
```

```python
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df["Day"]=df["Date"].dt.day
df["Month"]=df["Date"].dt.month
df["Year"]=df["Date"].dt.year
```
