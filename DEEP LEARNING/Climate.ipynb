```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
climate = pd.read_csv("climate_data(in).csv")
```

```python
climate
```

```python
climate.isna().sum()
```

```python
X = climate.drop('Energy Consumption', axis = 1) #Features
y = climate['Energy Consumption'] #Target
```

```python
X
```

```python
y
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

#scaler_y = StandardScaler()
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.2, random_state = 42)
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(64, input_dim = X_train.shape[1], activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1, activation = 'linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['mae'])
```

```python
model.fit(X_train, y_train, epochs = 50, batch_size = 10)
```

```python
loss = model.evaluate(X_test, y_test)
print(loss)
```

```python
from sklearn.metrics import r2_score
# Predict (scaled)
y_pred_scaled = model.predict(X_test)
print(f"Scaled output: {y_pred_scaled}")
# Inverse transform predictions
y_pred = scaler.inverse_transform(y_pred_scaled)
print(f"Normal Output: {y_pred}")
y_actual = scaler.inverse_transform(y_test)
print(f"Actual output: {y_pred_scaled}")

# R² score
r2 = r2_score(y_actual, y_pred)
print(f"R² Score: {r2:.2f}")
```
