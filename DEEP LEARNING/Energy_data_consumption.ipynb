```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
energy=pd.read_csv('predict_energy_consumption(in) (1).csv')
energy
```

```python
energy.columns
```

```python
x=energy[['temperature','humidity','wind_speed','solar_irradiance']]
y=energy['energy_consumption']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential(
    [Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
     Dense(32,activation='relu'),
     Dense(1)]
)
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
energy_model=model.fit(x_train_scaled,y_train,epochs=50,batch_size=32,validation_split=0.2)
```

```python
prediction=model.predict(x_test_scaled)
prediction
```

```python
test_loss,test_mae=model.evaluate(x_test_scaled,y_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')
```
