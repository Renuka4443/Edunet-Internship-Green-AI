```python
#temperature=[[75.2,77.1,74.5,79.3,82.6,81.2,77.8],[78,88,76,75,79,70,72]]
#explore the key attributes of numpy arrays ndim,shape,size,dtype,itemsize,and data
#demonstrate important numpy array function such as reshape,flatten and transponse
import numpy as np
temperature = [
    [75.2, 77.1, 74.5, 79.3, 82.6, 81.2, 77.8],
    [78, 88, 76, 75, 79, 70, 72]
]
# Convert to NumPy array
arr = np.array(temperature)
# Key Attributes
print("Original Array:\n", arr)
print("\nKey Attributes:")
print("Number of dimensions (ndim):", arr.ndim)
print("Shape (rows, columns):", arr.shape)
print("Total elements (size):", arr.size)
print("Data type (dtype):", arr.dtype)
print("Size of each element in bytes (itemsize):", arr.itemsize)
# NumPy Functions
print("\nReshaped to (7, 2):\n", arr.reshape(7, 2))
print("\nFlattened to 1D array:\n", arr.flatten())
print("\nTransposed array (rows <-> columns):\n", arr.transpose())
```
