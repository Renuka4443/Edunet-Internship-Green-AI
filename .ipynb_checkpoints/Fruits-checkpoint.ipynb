```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
data = {
    'Ripe': [1, 1, 1, 1, 0, 0],
    'Wormy': [0, 0, 1, 0, 0, 1],
    'Washed': [1, 0, 1, 0, 1, 1],
    'Eat': ['Yes', 'No', 'No', 'No', 'No', 'No']
}
```

```python
df=pd.DataFrame(data)
df
```

```python
#Define features and target
X=df[["Ripe","Wormy","Washed"]]
y=df["Eat"]
```

```python
#Train DT
model=DecisionTreeClassifier(criterion="entropy")
```

```python
model.fit(X,y)
```

```python
ripe=int(input("Whether fruit is ripe (0/1): "))
wormy=int(input("Whether fruit is wormy (0/1): "))
wash=int(input("Whether fruit is washed (0/1): "))
model.predict([[ripe,wormy,wash]])
```

```python
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names = X.columns, class_names= model.classes_, filled = True)
plt.title("Decision Tree : Should I eat the fruit?")
plt.show()
```
